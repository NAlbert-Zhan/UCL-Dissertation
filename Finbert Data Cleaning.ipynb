```python
import os
import shutil
import pandas as pd
```

```python
source_directory = 'C:/Users/lenovo/Desktop/UCL/Final dissertation/code/Data'  
destination_directory = 'C:/Users/lenovo/Desktop/UCL/Final dissertation/code/filtered data' 
os.makedirs(destination_directory, exist_ok=True)
```

```python
def extract_details_from_filename(filename):
    parts = filename.replace('.csv', '').split('-')
    CompanyTicker = parts[0]
    Year = int(float(parts[1]))  # Convert to float first, then to int to handle any decimal points
    return CompanyTicker, Year
```

```python
def extract_sentence_length(filename):
    csv_path = os.path.join(source_directory, filename)
    with open(csv_path, 'r', encoding='utf-8') as file:
        content = file.read()

    CSV_Length = content.count('\n')
    Status = 'TRUE' 
    return CSV_Length
```

```python
file_path = 'C:/Users/lenovo/Desktop/UCL/Final dissertation/code/Industrial Product.xlsx'
data = pd.read_excel(file_path)
data['Unnamed: 1'] = data['Unnamed: 1'].str.replace(' US Equity', '')
data.columns= ["Company", "Ticker", "Bi074", "Country", "2022", "2021", "2020", "2019", "2018", "2017", "2016", "2015"]
data=data.drop(0).reset_index(drop=True)

# Only select the rows with useful information
cols_to_check = ["2022", "2021", "2020", "2019", "2018", "2017", "2016", "2015"]
df = data.dropna(subset=cols_to_check, thresh=len(cols_to_check)-5)
```

```python
filtered_companyticker = df["Ticker"]
print(filtered_companyticker)
```

```python
for filename in os.listdir(source_directory):
    if filename.endswith('.csv'):  # Adjust the file extension if necessary
        CompanyTicker, Year = extract_details_from_filename(filename)
        CSV_Length = extract_sentence_length(filename)
        
        if 2015 <= Year <= 2022 and CSV_Length>=20:
            if CompanyTicker in filtered_companyticker.values:
                # Copy the file to the destination directory
                source_path = os.path.join(source_directory, filename)
                destination_path = os.path.join(destination_directory, filename)
                shutil.copy(source_path, destination_path)
                print(f"Copied {filename} to {destination_directory}")


print("File copying completed.")
```
